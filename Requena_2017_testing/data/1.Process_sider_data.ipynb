```python
# Needed modules
import pandas as pd
```

```python
# Import an Excel file
req_sider2 = pd.read_excel("./sider_2_requena_2017.xlsx")
print(req_sider2.shape)
print(req_sider2.dtypes)
req_sider2
```

```python
# Take what's only needed
req_sider2 = req_sider2[['Name', 'Coordinates']].copy()
print(req_sider2.shape)
print(req_sider2.dtypes)
req_sider2
```

```python
# Add colum 'sseqid', 'sstart', 'send'
req_sider2.loc[:, 'sseqid'] = req_sider2.loc[:, 'Coordinates'].str.split(':', expand=True)[0]
req_sider2.loc[:, 'sstart'] = req_sider2.loc[:, 'Coordinates'].str.split(':', expand=True)[1].str.split('-', expand=True)[0]
req_sider2.loc[:, 'send'] = req_sider2.loc[:, 'Coordinates'] .str.split(':', expand=True)[1].str.split('-', expand=True)[1]

# Let's see the final
print(req_sider2.shape)
print(req_sider2.dtypes)
req_sider2
```

```python
# Let's take only needed columns
req_sider2 = req_sider2[['sseqid', 'sstart', 'send']].copy()

# Let's transform 'sstart' and 'send' into 'int' using pandas
req_sider2['sstart'] = req_sider2['sstart'].astype(int)
req_sider2['send'] = req_sider2['send'].astype(int)

# Print the data types to confirm the conversion
print(req_sider2.dtypes)
req_sider2
```

```python
# Save the data
req_sider2.to_csv("./req_sider2.csv", index=False, sep=",", header=True)
```
