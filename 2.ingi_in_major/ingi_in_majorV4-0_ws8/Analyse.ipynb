````python
file_path = '/home/viskuit/Documents/Work_CBMSO/Bringaud_testing/2.ingi_in_major/ingi_in_majorV4-0_ws8/merged_sequences.csv'
file_to_compare = pd.read_csv(file_path, header=0, sep=',')

# Modify the "sseqid" column to keep only the part before the first underscore
file_to_compare['sseqid'] = file_to_compare['sseqid'].apply(lambda x: x.split('_')[0])
print(file_to_compare[['sseqid']].head())
# Modify the "sseqid" column to add a '.' before the first number
file_to_compare['sseqid'] = file_to_compare['sseqid'].apply(lambda x: x[:4] + '.' + x[4:])
print(file_to_compare[['sseqid']].head())
file_to_compare.to_csv(file_path, index=False, header=True, sep=',')
````
