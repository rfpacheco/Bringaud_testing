```python
import os
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord


def fasta_creator(data, fasta_output_path):
    matrix = []
    for index, sequence in data.iterrows():
        rec = SeqRecord(Seq(sequence['sseq']),
                        id=f"Seq_{index}_{sequence['sseqid']}",
                        description="Leishmania major"
                        )
        matrix.append(rec)
    SeqIO.write(matrix, fasta_output_path, "fasta")
    
data_path = "/2.ingi_in_major/ingi_in_majorV4-0_ws11/1.Analysing_software_results/final_yes_data.csv"
data_parent_path = os.path.dirname(data_path)
data_filename = os.path.basename(data_path)
data_save_fasta_path = os.path.join(data_parent_path, f"{data_filename}.fasta")

data_yes = pd.read_csv(data_path, sep=',', header=0)
fasta_creator(data=data_yes,
              fasta_output_path=data_save_fasta_path)
```
